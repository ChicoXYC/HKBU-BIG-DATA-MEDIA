# Brief Introduction 


TSPDT is called THEY SHOOT PICTURES, DON'T THEY?. TSPDT is an honest and evolving movie resource website. And I prefer to call it a database.


'1,000 Greatest Films' is their most well-known project, which is chosen by 4,343 film critics or film practitioners, and 3,302 magazine websites, a total of 8,797 exclusive lists. This version is their 13th and also the latest version. 

During this process of data analysis, we aim to figure out what kind of movies are most welcomed worldwide, and which director has earn most audience in those professionals and which country with its culture has dominate the film industry in the world.

```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('TSPTD 1000.csv')
```

```python
df
```

# Top 15 directors in the world


Trough the pic of top 15 directors, I found that French director Jean-Luc Godard has most films listes in this chart with 16 films. One thing I want to mention is that quality is not only criteria but quantity also matters.   

```python
fig = plt.figure(figsize=(14,7))
director_counts=df['Directors'].value_counts()[:15].sort_values(ascending=False).plot(
    kind='bar',color = '#46bc99',edgecolor = '#40b4e5')
plt.title('top 15 directors')
plt.xlabel('directors')
plt.ylabel('# of movies')
director_counts
```

```python
df['year'].min()
```

```python
year_counts = df['year'].value_counts()
year_counts
```

# Trend of the 'good' movies

You can see that there are several trend of movies that are most favored trough the pic of top 1000 movies by year.


The first trend is that there is a sharp increase in the pre 1950s, like the top 1 movie Citizen Kane is shot in this period . Anoter trend is that the period of 1955 - 1975 is the most productive period with many classics like Vertigo and 2001: A Space Odyssey. Meanwhile, so many film masters are active in those years, like Alfred Hitchcock, Stanley Kubrick, Jean-Luc Godard.

```python
plt.figure(figsize=(14, 7))
year_counts = year_counts.sort_index()
plt.plot(year_counts.index, year_counts.values,color='#46bc99')
ticks_index = range(0, 100, 10)
plt.xticks(ticks_index, year_counts.index[ticks_index])
plt.title('Top 1000 movies by year')
plt.xlabel('Year')
plt.ylabel('# of movies')
plt.annotate('A high productivity of good movies in 1955-1975', xy=(60, 20), xytext=(10, 20),
            arrowprops=dict(facecolor='black', width=2),
            )
plt.annotate('A sharp increase in the 1950s', xy=(42, 15), xytext=(16, 2),
            arrowprops=dict(facecolor='black', width=2),
            )
```

```python
region_counts = df['region'].value_counts()[:20]
region_counts
```

# Top 1000 movies by region

The chart of top 1000 movies by region shows that USA,France,Japan,UK and Italy, are the top 5 countries with most films listed in this chart, especially USA is 4 times more than the second country France, which I think can be illustrated by the dominent positions of these two contries in world culture, and they contribute most well-known films in the word with most developed film industry and system. 

```python
plt.figure(figsize=(20, 10))
region_counts = region_counts.sort_index()
plt.bar(region_counts.index, region_counts.values, color='#46bc99', edgecolor='#758be5')
ticks_index = range(0, 20, 1)
plt.xticks(ticks_index, region_counts.index[ticks_index])
plt.title('Top 1000 movies by region')
plt.xlabel('region')
plt.ylabel('# of movies')
```
